```python
categories = np.array(data.Category.unique())
print(categories)
breakfastMeals = data[data.Category == 'Breakfast']
beef_porkMeals = data[data.Category == 'Beef & Pork']
chicken_fishMeals = data[data.Category == 'Chicken & Fish']
saladMeals = data[data.Category == 'Salads']
snacks_sidesMeals = data[data.Category == 'Snacks & Sides']
desserts = data[data.Category == 'Desserts']
beverages = data[data.Category == 'Beverages']
coffee_tea = data[data.Category == 'Coffee & Tea']
smoothies_shakes = data[data.Category == 'Smoothies & Shakes']
```

```python
# How much calories does an average meal per category contribute to a male's daily caloric intake?
avgBreakfastCal = breakfastMeals.Calories.mean()
avgBeefPorkCal = beef_porkMeals.Calories.mean()
avgChickenFishCal = chicken_fishMeals.Calories.mean()
avgSaladCal = saladMeals.Calories.mean()
avgSnacksSidesCal = snacks_sidesMeals.Calories.mean()
avgDessertsCal = desserts.Calories.mean()
avgBeveragesCal = beverages.Calories.mean()
avgCoffeeTeaCal = coffee_tea.Calories.mean()
avgSmoothiesShakesCal = smoothies_shakes.Calories.mean()
avgCalsPerCategory = np.array([avgBreakfastCal, 
                              avgBeefPorkCal,
                              avgChickenFishCal,
                              avgSaladCal,
                              avgSnacksSidesCal,
                              avgDessertsCal,
                              avgBeveragesCal,
                              avgCoffeeTeaCal,
                              avgSmoothiesShakesCal])
print(avgCalsPerCategory)
np.ndarray.sort(avgCalsPerCategory, kind='heapsort')
y = np.linspace(0, 8, 9)
plt.barh(y, avgCalsPerCategory,height = 0.95, label = categories)
plt.yticks(y, categories)
plt.xlabel('Calories')
plt.style.use("ggplot")
plt.savefig("1.png",dpi=300)
```

```python
# Which meal has the most/least amount of sodium?
avgBreakfastSodium = breakfastMeals.Sodium.mean()
avgBeefPorkSodium = beef_porkMeals.Sodium.mean()
avgChickenFishSodium = chicken_fishMeals.Sodium.mean()
avgSaladSodium = saladMeals.Sodium.mean()
avgSnacksSidesSodium = snacks_sidesMeals.Sodium.mean()
avgDessertsSodium = desserts.Sodium.mean()
avgBeveragesSodium = beverages.Sodium.mean()
avgCoffeeTeaSodium = coffee_tea.Sodium.mean()
avgSmoothiesShakesSodium = smoothies_shakes.Sodium.mean()
avgSodiumPerCategory = np.array([avgBreakfastSodium, 
                              avgBeefPorkSodium,
                              avgChickenFishSodium,
                              avgSaladSodium,
                              avgSnacksSidesSodium,
                              avgDessertsSodium,
                              avgBeveragesSodium,
                              avgCoffeeTeaSodium,
                              avgSmoothiesShakesSodium])
np.ndarray.sort(avgSodiumPerCategory, kind='mergesort')
print(avgSodiumPerCategory)
y = np.linspace(0, 8, 9)
plt.barh(y, avgSodiumPerCategory,height = 0.95, label = categories,color='red')
plt.yticks(y, categories)
plt.xlabel('Sodium')
plt.savefig("2.png",dpi=300)
```

```python
data['percent fat'] = 100*data['Calories from Fat']/data.Calories

g = data[['Item','percent fat']].sort_values('percent fat', ascending = False)

g = g[g['percent fat']>50]


fig, ax = plt.subplots(figsize=(5,15))
sns.stripplot(ax = ax,data = g, x = 'percent fat', y = 'Item', s = 10, color = 'grey')

ax.yaxis.grid(True)

ax.set_xlabel('Percent Calories from Fat')
plt.savefig("4.png",dpi=300)
```

```python
order = data.groupby('Category').agg({'percent fat':'mean'}).sort_values('percent fat', ascending = False).index

fig, ax = plt.subplots(figsize=(8,5))

sns.factorplot(data = data, x = 'Category',y = 'percent fat', ax = ax, order = order)

ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)

ax.set_ylabel('Mean Percent Calories from Fat')
plt.savefig("5.png",dpi=300)
```

```python
measures = ['Calories', 'Total Fat', 'Cholesterol','Sodium', 'Sugars', 'Carbohydrates']

for m in measures:   
    plot = sns.swarmplot(x="Category", y=m, data=data)
    plt.setp(plot.get_xticklabels(), rotation=45)
    plt.title(m)
plt.savefig("5.png",dpi=300)
```

```python
p=sns.boxplot(data= data, x = 'Category',y = 'Dietary Fiber')
plt.tight_layout
p.set_xticklabels(p.get_xticklabels(),rotation = 45)
plt.savefig("6.png",dpi=300)
```

```python
def barplotter(grouped):
    item = grouped["Item"].sum()
    item_list = item.sort_index()
    item_list = item_list[-20:]
    #Sizing the image canvas
    plt.figure(figsize=(8,9))
    #To plot bargraph
    g = sns.barplot(item_list.index,item_list.values)
    labels = [aj.get_text()[-40:] for aj in g.get_yticklabels()]
    g.set_yticklabels(labels)
```

```python
alpha = data.groupby(data["Carbohydrates"])
barplotter(alpha)
plt.show()
```

```python
#Plot for protein
beta = data.groupby(data["Protein"])
barplotter(beta)
plt.show()
```

```python
#Plot for calories
gamma = data.groupby(data["Calories"])
barplotter(gamma)
plt.show()
```

```python
#Plot for Dietary Fiber(% Daily Value)
delta = data.groupby(data["Dietary Fiber (% Daily Value)"])
barplotter(delta)
plt.show()
```
