```python
import json
f = open("landline_subscription.txt", 'r')
j = f.read()
dictonary = json.loads(j)
```

```python
dictonary
```

```python
x = []
y = []
for d in dictonary:
    x.append(d['year'])
    y.append(d['value'])
```

```python
x
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```
