```python
flavours = whiskey.iloc[:,2:14] #extracting data from table
```

```python
corr = pd.DataFrame.corr(flavours)
```

```python
print(corr)
```

```python
plt.figure(figsize=(10,10))
plt.pcolor(corr)
plt.colorbar()
plt.show()
```

```python
corr_whiskey = pd.DataFrame.corr(flavours.transpose())
plt.figure(figsize=(10,10))
plt.pcolor(corr_whiskey)
plt.colorbar()
plt.show()
```
