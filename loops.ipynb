# Loops

## while (some conditions)

```python
%%file wile_loop.y
n = 1
while n <= 5:
    print(n)
    n = n + 1
print("outside the loop")
```

### For - loop

```python
fruits = ['apples', 'oranges', 12.5, 'kiwi', 'strawberry', 'mango', 'banana', 'grapes']
```

```python
fruits
```

```python
for i in fruits:
    print(i)
```
