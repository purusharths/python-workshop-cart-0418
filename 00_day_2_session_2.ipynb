## Sub plots

```python
x = np.linspace(0,2*np.pi,100)
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.subplot(2,2,3)
plt.plot(x, np.tan(x))
plt.subplot(2,2,4)
plt.plot(x,x)
plt.show()
```
