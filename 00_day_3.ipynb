## KDE Plot

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
setosa = df[df['species'] == 'setosa']
virginica = df[df['species'] == 'virginica']
```

```python
virginica.head(n=1)
```

```python
plt.scatter(setosa.sepal_width, setosa.sepal_length, color="r")
plt.scatter(virginica.sepal_width, virginica.sepal_length, color='b')
```

```python
f, ax = plt.subplots(figsize=(8, 8))

sns.kdeplot(setosa.sepal_width, setosa.sepal_length,
                  cmap="Reds", shade=False)

# sepal_width vs sepal_length of virginica
sns.kdeplot(virginica.sepal_width, virginica.sepal_length,
                 cmap="Blues", shade=True, shade_lowest=False)


ax.text(2.5, 8.2, "virginica", size=16)
ax.text(3.8, 4.5, "setosa", size=16)
```

## sampling

```python
df_female = df[df['sex'] == 'female'] # df_female => population
```

```python
df_female.shape
```

```python
sample = df_female.sample(frac=0.1) # sample is 10% of df_female
```

```python
sample.head()
```

```python
sample_mean = sample['weight'].mean()
```

```python
sample_mean
```

## Sampling Distribution of Sample Means

```python
independent_means = []
for i in range(1000):
    sample = df_female.sample(frac=0.1)
    sample_mean = sample['weight'].mean()
    independent_means.append(sample_mean)
```

```python
sns.distplot(independent_means)
```

$$\sigma_{x'} \approx \frac{s}{\sqrt{n}}$$

```python
sample_variance = sample['weight'].std()/np.sqrt(len(sample))
```

```python
sample_variance
```

```python
def zscore(val):
    return stats.norm.ppf((1+val)/2)
```

```python
z = zscore(0.95)
```

```python
lower_limit = sample_mean - z*sample_variance
upper_limit = sample_mean + z*sample_variance

lower_limit, upper_limit
```
